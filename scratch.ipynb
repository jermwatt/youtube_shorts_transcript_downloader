```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)
```

```python
# Wait for the "More actions" button to be clickable and click it
more_actions_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='More actions']")))
more_actions_button.click()

# Wait for the "Show transcript" option to be visible and click it
show_transcript_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//ytd-menu-service-item-renderer[contains(., 'Show transcript')]")))
show_transcript_button.click()
```
